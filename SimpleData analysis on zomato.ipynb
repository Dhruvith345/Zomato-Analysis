```python
# Assumption:1 Average cost affects the rating:
avg_cost1=zomato[zomato['rate (out of 5)']<3.5]
avg_cost1['rate (out of 5)']
```

```python
avg_cost2=zomato[zomato['rate (out of 5)']>4.4]
avg_cost2
```
